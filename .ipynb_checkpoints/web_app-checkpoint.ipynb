```python
from pymongo import MongoClient

import dash
import dash_core_components as dcc 
import dash_html_components as html
from dash.dependencies import Input, Output

#Récupération des données############################################################################################
#Lien entre la base mongo et le programme
client = MongoClient()

#accès à nos données
db_Manga = client.MangaDB     
collection_Manga = db_Manga['mangaCollection']

#création d'une liste de dictionnaire contenant les différents genres de manga
liste_genre = [{
            'label':"Tout les genres",
            'value':"all"
            }]
cursor = collection_Manga.find()
for document in cursor:
    d = {
        'label':document['genre'],
        'value':document['genre']
    }
    if d not in liste_genre:
        liste_genre.append(d)

#création d'une liste contenant une ligne pour chaque manga
cursor = collection_Manga.find()

liste_livre = [html.Tr([html.Td(children=html.Strong("titre")),
            html.Td(children=html.Strong("synopsis")),
            html.Td(children=html.Strong("genre")),
            html.Td(children=html.Strong("episodes")),
            html.Td(children=html.Strong("note"))])]

for document in cursor:
    line = html.Tr([html.Td(children=document["titre"]),
            html.Td(children=document["synopsis"]),
            html.Td(children=document["genre"]),
            html.Td(children=document["episodes"]),
            html.Td(children=document["note"])])
    liste_livre.append(line)

    
    
#Application web#################################################################################    
#style général de la page web
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

# Création de la webapp
app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

app.layout = html.Div(children=[
    
                        html.H1(children='Recherchez un manga dans la barre de recherche',
                                    style={'textAlign': 'center', 'color': '#7FDBFF'}),
                        
                        dcc.Input(id='barre_recherche', value='enter manga title', type='text'),
                        html.Table(id = "specific_manga",children=[html.Tr([html.Td(children=html.Strong("titre")),
                                                                    html.Td(children=html.Strong("synopsis")),
                                                                    html.Td(children=html.Strong("genre")),
                                                                    html.Td(children=html.Strong("episodes")),
                                                                    html.Td(children=html.Strong("note"))])]),
                        
                        html.H1(children='Selectionnez un genre de manga',
                                    style={'textAlign': 'center', 'color': '#7FDBFF'}),

                        dcc.Dropdown(id = "genre_selection",options=liste_genre,
                                    value='all'),
    
                        html.Table(id = "manga_table",children=liste_livre)

])
#CALLBACK#######################################################################################
@app.callback(
    Output('manga_table', 'children'),
    Input('genre_selection', 'value'))
def callback_genre(x):
    
    liste_livre = [html.Tr([html.Td(children=html.Strong("titre")),
            html.Td(children=html.Strong("synopsis")),
            html.Td(children=html.Strong("genre")),
            html.Td(children=html.Strong("episodes")),
            html.Td(children=html.Strong("note"))])]
    
    cursor = collection_Manga.find()
    
    for document in cursor:
        if document["genre"]==x or x == "all":
            line = html.Tr([html.Td(children=document["titre"]),
                    html.Td(children=document["synopsis"]),
                    html.Td(children=document["genre"]),
                    html.Td(children=document["episodes"]),
                    html.Td(children=document["note"])])
            liste_livre.append(line)
    return liste_livre


@app.callback(
    Output('specific_manga', 'children'),
    Input('barre_recherche', 'value'))
def callback_titre(x):
    
    liste_livre = [html.Tr([html.Td(children=html.Strong("titre")),
            html.Td(children=html.Strong("synopsis")),
            html.Td(children=html.Strong("genre")),
            html.Td(children=html.Strong("episodes")),
            html.Td(children=html.Strong("note"))])]
    
    cursor = collection_Manga.find()
    for document in cursor:
        if document["titre"].upper().startswith(x.upper()):
            line = html.Tr([html.Td(children=document["titre"]),
                    html.Td(children=document["synopsis"]),
                    html.Td(children=document["genre"]),
                    html.Td(children=document["episodes"]),
                    html.Td(children=document["note"])])
            liste_livre.append(line)
    return liste_livre



app.run_server()
```
